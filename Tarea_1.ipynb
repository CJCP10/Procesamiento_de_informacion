Crear un notebook para preprocesar el texto y convertirlo a minúsculas, quitar acentos y los siguientes caracteres:

";:,.\\-\"'/()[]¿?¡!{}~<>|«»-—’\t\n\r"

```python
import unicodedata

def preprocesar_texto(texto):
    texto = texto.lower()
    # Normaliza el texto a NFKD y elimina los acentos
    texto = unicodedata.normalize('NFKD', texto)
    texto = texto.encode('ASCII', 'ignore').decode('utf-8')

    caracteres_a_quitar = ";:,.\\-\"'/()[]¿?¡!{}~<>|«»-—’\t\n\r"
    for caracter in caracteres_a_quitar:
        texto = texto.replace(caracter, "")

    return texto
```

Calcular las  frecuencias a nivel de un carácter, de los documentos text_1 al text_5  y generar los histogramas para cada archivo. Para generar las gráficas, ordenar los caracteres de acuerdo a su frecuencia; de mayor a menor frecuencia en el histograma.

```python
def calcular_frecuencias(texto):
    frecuencias = {}
    for caracter in texto:
        if caracter in frecuencias:
            frecuencias[caracter] += 1
        else:
            frecuencias[caracter] = 1
    return frecuencias
```

```python
import matplotlib.pyplot as plt

# Datos de frecuencias
frecuencias = calcular_frecuencias(text)

# Ordenamos de mayor a menor las frecuencias
frecuencias = dict(sorted(frecuencias.items(), key=lambda item: item[1], reverse=True))
# Separar claves y valores
caracteres = list(frecuencias.keys())
valores = list(frecuencias.values())

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.bar(caracteres, valores, color='skyblue')

# Personalizar el gráfico
plt.title('Frecuencia de Caracteres')
plt.xlabel('Caracteres')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)  # Rotar etiquetas en eje x para mejor visualización

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
